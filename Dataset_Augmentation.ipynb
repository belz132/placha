```python
import pandas as pd

df_train = pd.read_csv('train_preprocess_labeled.tsv', sep='\t')
df_valid = pd.read_csv('valid_preprocess_labeled.tsv', sep='\t')
df = df_train.append(df_valid, ignore_index=True)
df_test = pd.read_csv('test_preprocess_labeled.tsv', sep='\t')
df = df.append(df_test, ignore_index=True)
```

```python
positive_df = df[df['label']=='positive']
negative_df = df[df['label']=='negative']
neutral_df = df[df['label']=='neutral']
```

```python
positive_list = positive_df.text.to_list()
positive_list[0:5]
```

```python
negative_list = negative_df.text.to_list()
negative_list[0:5]
```

```python
neutral_list = neutral_df.text.to_list()
neutral_list[0:5]
```

```python
from googletrans import Translator

translator = Translator(service_urls = ["translate.googleapis.com"])

# translate to English
to_en = translator.translate(positive_list[2], dest = 'en')
# translate back to Indonesian
to_id = translator.translate(to_en.text, dest = 'id').text
to_id
```

```python
# Generating new negative text, from existing negative text, by translating it to English and back to Indonesian
# ~20:xxpm-
translated_negative_list = []

for data in negative_list[:]:
    
    # translate to English and back to Indonesian
    to_en = translator.translate(data, dest = 'en')
    to_id = translator.translate(to_en.text, dest = 'id').text
    translated_negative_list.append(to_id)
    
translated_negative_list
```

```python
translated_negative_df = pd.DataFrame({'text': translated_negative_list,'label': 'negative'})
augmented_negative_df = pd.concat([negative_df, translated_negative_df])

augmented_negative_df['label'].value_counts()
```
